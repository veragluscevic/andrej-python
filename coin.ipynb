Calculating the probabilities of coin-flip outcomes

Python imports we'll need later...

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

%matplotlib inline


```

```python
def probability_number(ntot, nhead):
    bi = sp.special.binom(ntot, nhead, out=None)
    return bi
    
```

```python
ntot = 10
nhs = np.arange(0, ntot + 1)
probs = probability_number(ntot, nhs)
plt.plot(nhs, probs)
```

```python

ntot = 1000
nhs = np.arange(0, ntot + 1)
probs = probability_number(ntot, nhs)
plt.plot(nhs, probs)
```
